### Generate matrices A, with random Gaussian entries, B, a Toeplitz matrix, where A ∈ Rn×m and B ∈ Rm×m, for n = 200, m = 500.

```python
from scipy.linalg import toeplitz
import numpy as np


'''
    n   m
A = 200*500 --> random gaussian entries
B = 500*500 --> toeplitz matrix
'''
n=200
m=500
A = np.random.normal(size=(n,m))
B = toeplitz(np.random.rand(500),np.random.rand(500))

print(A)
```

### 9.1: Write a function that computes A(B −λI) for any λ.

```python
#print(len(B))

def compute(A,B,λ):

    I = np.identity(len(B))
    λI = np.dot(λ,I)
 
    return np.dot(A, B-λI)

compute(A,B,500)
```

### Exercise 9.6: Nearest neighbor Write a function that takes a value z and an array A and finds the element in A that is closest to z. The function should return the closest value, not index. Hint: Use the built-in functionality of Numpy rather than writing code to find this value manually. In particular, use brackets and argmin.

```python
A = np.array([
             [1, 2, 3], 
             [4, 25, 6],
             [7, 10, 9]
             ]
            )
def nearestNeighbor(z,A):
    #turn array to 1d array 
    #turn A-z array to 1d array and get index of minimum element (index of minimum --> (A-z))
    #get element with found index 
    #dimension of array is does not matter in this solution
    
    return A.flatten()[np.absolute(A-z).flatten().argmin()]
    

nearestNeighbor(11,A)
```
